```python
import sys
import os

# Ensure app/ folder is in Python path
sys.path.append(os.path.abspath("../app"))

from backend import get_intent, get_response, extract_role_from_text
```

```python
tests = [
    "What skills are needed for Data Scientist?",
    "Give me the roadmap for Backend Developer",
    "Project ideas for Cloud Engineer",
    "Explain the role of Data Analyst",
    "Difference between Data Scientist and Data Analyst",
    "Give me resume tips",
    "What skills do I need?",
    "Give me a roadmap",
    "Projects please",
    "hello"
]
```

```python
results = []

for query in tests:
    intent = get_intent(query)
    role = extract_role_from_text(query)
    response = get_response(query, intent)

    results.append({
        "query": query,
        "intent": intent,
        "role": role,
        "response_preview": response[:200] + "..."  # shorten long responses
    })

results
```

```python
for r in results:
    print("="*60)
    print("Query:", r["query"])
    print("Intent:", r["intent"])
    print("Role:", r["role"])
    print("Response:", r["response_preview"])
    print()
```
